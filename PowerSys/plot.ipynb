```python
import matplotlib.pyplot as plt
from sympy import *

x1reqr=[]
na=[]
n=symbols('n')
req=symbols('r_eq')
Dm=symbols('D_m')
x1=0.1445*log(Dm/req)+0.0157/n
x1Dm=x1.subs(Dm,0.4)
x1req=x1Dm.subs(req,0.5)
i=0
for i in range(60):
    x1reqr.append( x1req.subs(n,(i+1)/10))
    na.append((i+1)/10)
plt.plot(na,x1reqr)
x1reqr
#plt.plot([1, 2, 3, 4], n)
```
